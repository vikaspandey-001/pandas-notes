# Merge

```python
import pandas as pd
var1= pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2= pd.DataFrame({"A":[1,2,3,4],"C":[21,22,23,24]})
pd.merge(var1,var2)
```

```python
import pandas as pd
var1= pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2= pd.DataFrame({"A":[1,2,3,4],"C":[21,22,23,24]})
pd.merge(var1,var2,on="A")     # "on" is used to show common key in both dataframe. Although it merges without without "on" only in new versions of jupyter.
```

```python
import pandas as pd
var1= pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2= pd.DataFrame({"A":[1,2,3,5],"C":[21,22,23,24]})
pd.merge(var1,var2,how="inner")        #it merges only common part (or intersection) but the condition is both the dataframes have same length
```

```python
import pandas as pd
var1= pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2= pd.DataFrame({"A":[1,2,3,5],"C":[21,22,23,24]})
pd.merge(var1,var2,how="outer")        #it merges all ( or universal) but the condition is both the dataframes have same length
```

```python
import pandas as pd
var1= pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2= pd.DataFrame({"A":[1,2,3,5],"C":[21,22,23,24]})
pd.merge(var1,var2,how="left")        #it merges on left but the condition is both the dataframes have same length
```

```python
import pandas as pd
var1= pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2= pd.DataFrame({"A":[1,2,3,5],"C":[21,22,23,24]})
pd.merge(var1,var2,how="right")        #it merges on right but the condition is both the dataframes have same length
```

```python
import pandas as pd
var1= pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2= pd.DataFrame({"A":[1,2,3,5],"C":[21,22,23,24]})
pd.merge(var1,var2,how="outer",indicator=True)        #it merges all ( or universal) but the condition is both the dataframes have same length
```

```python
import pandas as pd
var1= pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2= pd.DataFrame({"A":[1,2,3,5],"C":[21,22,23,24]})
pd.merge(var1,var2,how="inner",indicator=True)        #it merges only common part (or intersection) but the condition is both the dataframes have same length
```

```python
import pandas as pd
var1= pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2= pd.DataFrame({"A":[1,2,3,5],"B":[21,22,23,24]})
pd.merge(var1,var2,left_index=True,right_index=True)
```

```python
import pandas as pd
var1= pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2= pd.DataFrame({"A":[1,2,3,5],"B":[21,22,23,24]})
pd.merge(var1,var2,left_index=True,right_index=True,suffixes=("name","id"))
```

```python
import pandas as pd
var1= pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2= pd.DataFrame({"A":[1,2,3,5],"B":[21,22,23,24]})
pd.merge(var1,var2,left_index=True,right_index=True,preffixes=("name","id"))
```

There is no argument named prefixes in pandas

# Concat

```python
import pandas as pd
var1= pd.Series([1,2,3,4],[11,12,13,14])
var2= pd.Series([1,2,3,5],[21,22,23,24])
pd.concat([var1,var2]) 
```

```python
import pandas as pd
var1= pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2= pd.DataFrame({"A":[1,2,3,5],"B":[21,22,23,24]})
pd.concat([var1,var2]) 
```

```python
import pandas as pd
var1= pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2= pd.DataFrame({"C":[1,2,3,5],"D":[21,22,23,24]})
pd.concat([var1,var2]) 
```

```python
import pandas as pd
var1= pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2= pd.DataFrame({"A":[1,2,3,5],"B":[21,22,23,24]})
pd.concat([var1,var2],axis=1) 
```

```python
import pandas as pd
var1= pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2= pd.DataFrame({"A":[1,2,3,5],"B":[21,22,23,24]})
pd.concat([var1,var2],axis=0) 
```

```python
import pandas as pd
var1= pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2= pd.DataFrame({"A":[1,2,3,5],"B":[21,22,23,24]})
pd.concat([var1,var2],axis=1,suffixes="Hello_") 
```

concat has no argument suffixes.

```python
import pandas as pd
var1= pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2= pd.DataFrame({"A":[1,2],"C":[21,22]})
pd.concat([var1,var2],axis=1,join="outer")       # it shows all concated data
```

```python
import pandas as pd
var1= pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2= pd.DataFrame({"A":[1,2],"C":[21,22]})
pd.concat([var1,var2],axis=1,join="inner")       # it shows only intersection concated data i.e it removes NaN from data     .
```

```python
import pandas as pd
var1= pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2= pd.DataFrame({"A":[1,2,3,5],"C":[21,22,23,24]})
pd.concat([var1,var2],axis=1,keys=["V","F"])  
```

```python
import pandas as pd
var1= pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2= pd.DataFrame({"A":[1,2,3,5],"C":[21,22,23,24]})
pd.concat([var1,var2],axis=0,keys=["V","F"])    
```
