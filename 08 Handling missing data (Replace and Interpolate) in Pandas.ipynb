# Interpolate :The interpolate() method replaces the NULL values based on a specified method. Only work for numeric data not with alphabetic data.

```python
import pandas as pd
var=pd.read_csv("C:\\Users\\Vikas Pandey\\Downloads\\student.1.csv")          
var
```

```python
import pandas as pd
var_12=pd.read_csv("C:\\Users\\Vikas Pandey\\Downloads\\student.1.csv")          
var_12.interpolate() 
```

#### interpolate(method=("_")) : {linear, time, index, values, nearest, zero, slinear, quadratic, cubic, barycentric, krogh, polynomial, spline, piecewise_polynomial, from_derivatives, pchip, akima}

```python
import pandas as pd
var_13=pd.read_csv("C:\\Users\\Vikas Pandey\\Downloads\\student.1.csv")          
var_13.interpolate(method="linear")
```
