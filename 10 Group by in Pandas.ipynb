# Group by 

Group by is used to merge duplicate data and perform mathematical operations.

```python
import pandas as pd
var=pd.DataFrame({"Name":["a","b","c","d","a","b","c","d","a","b","c","d","a","b","c","d"],
                     "S_1":[12,13,15,18,8,10,36,58,21,31,25,10,100,25,92,18],
                    "S_2":[15,65,85,15,35,49,6,4,75,1,3,5,6,8,215,5]})
var
```

```python
import pandas as pd
var=pd.DataFrame({"Name":["a","b","c","d","a","b","c","d","a","b","c","d","a","b","c","d"],
                     "S_1":[12,13,15,18,8,10,36,58,21,31,25,10,100,25,92,18],
                    "S_2":[15,65,85,15,35,49,6,4,75,1,3,5,6,8,215,5]})
var_new=var.groupby("Name")
for x,y in var_new:
    print(x)
    print(y)
```

```python
import pandas as pd
var=pd.DataFrame({"Name":["a","b","c","d","a","b","c","d","a","b","c","d","a","b","c","d"],
                     "S_1":[12,13,15,18,8,10,36,58,21,31,25,10,100,25,92,18],
                    "S_2":[15,65,85,15,35,49,6,4,75,1,3,5,6,8,215,5]})
var_new=var.groupby("Name")
var_new.get_group("a")        # it gives only data of "a"
```
