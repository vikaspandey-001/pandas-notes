```python
import pandas as pd
var1= pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]},index=["a","b","c","d"])
var2= pd.DataFrame({"A":[1,2],"D":[23,24]},index=["a","b"])
var1.append(var2)
```

```python
import pandas as pd
var1= pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})
var2= pd.DataFrame({"A":[1,2],"D":[23,24]})
var1.append(var2)
```
