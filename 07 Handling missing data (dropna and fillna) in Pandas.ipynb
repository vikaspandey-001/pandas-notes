```python
import pandas as pd
var=pd.read_csv("C:\\Users\\Vikas Pandey\\Downloads\\student.csv")          
var_1=var.dropna()          #it will drop whole row if any value is found to be NaN
print(var_1)         
print(var_1.index)          #returns index of remaining data in list form
print(var_1.count())
```

```python
import pandas as pd
var=pd.read_csv("C:\\Users\\Vikas Pandey\\Downloads\\student.csv")          
var_2=var.dropna(axis=1)          #it will drop whole column if any column having atleast one NaN value
print(var_2)         
print(var_2.index)                #returns index of remaining data in list form
print(var_2.count())
```

```python
import pandas as pd
var=pd.read_csv("C:\\Users\\Vikas Pandey\\Downloads\\student.csv")          
var_3=var.dropna(axis=0)          #it will drop whole row if any row having atleast one NaN value                               
print(var_3)                      # or we can use only dropna()
print(var_3.index)                #returns index of remaining data in list form
print(var_3.count())
```

```python
import pandas as pd
var=pd.read_csv("C:\\Users\\Vikas Pandey\\Downloads\\student.csv")          
var_4=var.dropna(how="any")          #it will drop whole row if all the values of that row is Nan    
print(var_4)                         #or we can use dropna()
print(var_4.index)                   #returns index of remaining data in list form
print(var_4.count())
```

```python
import pandas as pd
var=pd.read_csv("C:\\Users\\Vikas Pandey\\Downloads\\student.csv")          
var_5=var.dropna(how="all")          #does nothing (in suspesion) #it will drop whole row if any row having atleast one NaN value                               
print(var_5)                         
print(var_5.index)                   #returns index of remaining data in list form
print(var_5.count())                 
```

```python
import pandas as pd
var=pd.read_csv("C:\\Users\\Vikas Pandey\\Downloads\\student.csv")          
var_6=var.dropna(subset=["gender"])  #it will drop only row of gender column having NaN value      
print(var_6)                         
print(var_6.index)                   #returns index of remaining data in list form
print(var_6.count())                 
```

### Fillna

```python
import pandas as pd
var=pd.read_csv("C:\\Users\\Vikas Pandey\\Downloads\\student.csv")          
var_10=var.fillna("NOT")   #it will fill all NaN with "NOT"
var_10
```

```python
import pandas as pd
var=pd.read_csv("C:\\Users\\Vikas Pandey\\Downloads\\student.csv")          
var_11=var.fillna({"id":"NOT","name":"Chutiya hai","class":"Padhta hi nhi","mark":"Nalayak","gender":"Pata nhi"})
var_11                 # it will replace NaN value according to their values in dictionary
```

```python
import pandas as pd
var=pd.read_csv("C:\\Users\\Vikas Pandey\\Downloads\\student.csv")          
var_12=var.fillna(method="ffill")   #it will fill all NaN same with repect to its forward value
var_12
```

```python
import pandas as pd
var=pd.read_csv("C:\\Users\\Vikas Pandey\\Downloads\\student.csv")          
var_13=var.fillna(method="bfill")   #it will fill all NaN same with repect to its backward value
var_13
```

```python
import pandas as pd
var=pd.read_csv("C:\\Users\\Vikas Pandey\\Downloads\\student.csv")          
var_14=var.fillna(method="ffill",axis=0)   #it will fill all NaN same with repect to its forward row value
var_14
```

```python
import pandas as pd
var=pd.read_csv("C:\\Users\\Vikas Pandey\\Downloads\\student.csv")          
var_15=var.fillna(method="ffill",axis=1)   #it will fill all NaN same with repect to its forward column value
var_15
```

```python
import pandas as pd
var=pd.read_csv("C:\\Users\\Vikas Pandey\\Downloads\\student.csv")          
var_16=var.fillna(method="bfill",axis=0)   #it will fill all NaN same with repect to its backward row value
var_16
```

```python
import pandas as pd
var=pd.read_csv("C:\\Users\\Vikas Pandey\\Downloads\\student.csv")          
var_17=var.fillna(method="bfill",axis=1)   #it will fill all NaN same with repect to its backward column value
var_17                          #Yaha gender ki NaN value pe koi farak nhi padega  kyunki gender ke backward me koi column ich nhi hai "
```
